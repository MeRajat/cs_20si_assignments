```python
import tensorflow as tf 
```

```python
a = tf.constant(2)
b = tf.constant(3)
c = tf.add(a, b)
```

```python
with tf.Session() as sess:
    # this saves graph info in graphs directory for tensorboard to process 
    writer = tf.summary.FileWriter('./graphs', sess.graph)
    print(sess.run(c))
    
    writer.close()
```

## Constants

```python
a = tf.constant([2,2], name = "a")
b = tf.constant([[0,1], [2,3]], name = "b")
x = tf.add(a,b, name = "add")
y = tf.multiply(a,b, name = "mul")

with tf.Session() as sess:
    writer = tf.summary.FileWriter("./graphs", sess.graph)
    x,y = sess.run([x,y])
    print(x,y)
    writer.close()
```

```python
my_const = tf.constant([1.0, 2.0], name = "my_const")
with tf.Session() as sess:
    print(sess.graph.as_graph_def())
```

#### Why Constants are bad 
    Constants are stored inside graph defination (see o/p of sess.graph.as_graph_def()) 
    This makes loading graphs expensive when constants are big
    
#### Solution 
    Only use constants for primitve types.
    Use variables or readers for more data that requires more memory.

## Variables

```python
a = tf.Variable(tf.zeros([784, 10]))
```

#### Why tf.constant but tf.Variable and not tf.variable
    tf.Variable is a class but tf.constant is an op 
    
    x.initializer # init op 
    x.value() # read op 
    x.assign(...) # write op 
    more
    

##### initalize your variables
    easiest way is initializing all variables at once. 
    
    init = tf.global_variables_initalizer()
    with tf.Session() as sess:
        sess.run(init)
    
#### Initalize only subset of variables
    init_ab = tf.variables_initalizer([a,b], name = "init_ab")
    
#### Initalize a single variable 

    W = tf.Variable(tf.zeros([784, 10))
    with tf.Session() as sess:
        sess.run(W.iinitalizer)
